```python
Задание: Итератор по операциям клиента банка
Есть клиент банка, у которого есть список операций по счёту. Каждая операция представлена словарём с типом операции и суммой. Тип операции может быть "deposit" (пополнение) или "withdraw" (списание). Аналитику нужно уметь проходить по операциям по одной, используя итератор, и на основе этого считать итоговый баланс клиента.
Первый класс — ClientOperations.
Он должен:
принимать список операций в конструкторе
хранить этот список
быть итерируемым объектом
реализовывать метод __iter__(), который возвращает новый итератор при каждом вызове

Второй класс — ClientOperationsIterator.
Он должен:
принимать список операций
хранить текущую позицию перебора
реализовывать метод __next__()
возвращать операции по одной
выбрасывать StopIteration, когда операции заканчиваются


Формат операций:
operations = [
{"type": "deposit", "amount": 1000},
{"type": "withdraw", "amount": 300},
{"type": "deposit", "amount": 500},
]


После реализации классов код должен работать следующим образом:
client_ops = ClientOperations(operations)

balance = 0

for op in client_ops:
if op["type"] == "deposit":
balance += op["amount"]
else:
balance -= op["amount"]

print(balance) # ожидаемый результат: 1200
```

```python
class ClientOperations:
    def __init__(self, operations):
        self.operations=operations

    def __iter__(self):
        return ClientOperationsIterator(self.operations)
    
class ClientOperationsIterator:
   def __init__(self, operations):
        self.operations=operations
        self.index=0

   
   def __next__(self):
        if self.index>=len(self.operations):
            raise StopIteration
        
        operation = self.operations[self.index]
        self.index+=1
        return operation

operations = [
{"type": "deposit", "amount": 1000},
{"type": "withdraw", "amount": 300},
{"type": "deposit", "amount": 500},
]

operations = ClientOperations(operations)

for ops in operations:
    print(ops['amount'])

client_ops = ClientOperations(operations)

balance = 0

for ops in client_ops:
    
    if ops["type"] == "deposit":
        balance += ops["amount"]
    else:
        balance -= ops["amount"]

print(balance)

    
 
```
