```python
Условие
Создайте базовый класс Employee.
У класса должно быть:
защищённый атрибут _salary
конструктор __init__(salary)
метод get_salary(), который возвращает зарплату
метод get_role(), который возвращает строку "Employee"

Создайте дочерний класс Manager.
Класс должен:
наследоваться от Employee
переопределять метод get_role()
возвращать строку "Manager"
иметь метод get_salary(), который возвращает зарплату с бонусом +1000

Создайте дочерний класс Developer.
Класс должен:
наследоваться от Employee
переопределять метод get_role()
возвращать строку "Developer"
возвращать зарплату без изменений

Создайте список сотрудников и выведите информацию о каждом:
employees = [
Employee(3000),
Manager(5000),
Developer(4000)
]
Для каждого объекта выведите:
роль (get_role())
зарплату (get_salary())

Пример ожидаемого вывода
Role: Employee, Salary: 3000
Role: Manager, Salary: 6000
Role: Developer, Salary: 4000


Подсказки
используйте _salary (инкапсуляция)
используйте переопределение методов (полиморфизм)
вызывайте одинаковые методы у разных объектов
```

```python
class Employee:
    def __init__(self, salary):
        self.salary=salary

    
def get_salary(self):
    return self.salary

def get_role(self):
    return "Employee"

    
class Manager(Employee):
   def get_role(self):
        return "Manager"
   def get_salary(self):
        return self._salary + 1000

class Developer(Employee):
    def get_role(self):
        return "Developer"
   
employees = [
    Employee(3000),
    Manager(5000),
    Developer(4000)
]

for emp in employees:
     print(f"Role: {emp.get_role()}, Salary: {emp.get_salary()}")

```
