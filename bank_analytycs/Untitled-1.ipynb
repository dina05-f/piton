```python
#HW 1 OOP
# Back

#ЗАДАНИЕ: Персонаж в игре

class Character:
    def __init__(self, name, hp):
        self.name = name
        self._hp = 0
        self.hp = hp  # используем сеттер для проверки границ
```

```python
@property
def hp(self):
    return self._hp
```

```python
@hp.setter
def hp(self, value):
        if value < 0:
            self._hp = 0
        elif value > 100:
            self._hp = 100
        else:
            self._hp = value
```

```python
def take_damage(self, amount):
        self.hp -= amount
```

```python
def heal(self, amount):
        self.hp += amount
```

```python
def status(self):
        print(f"{self.name}: HP = {self.hp}")
```
