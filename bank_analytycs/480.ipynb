```python
Задание 1. ООП (наследование + инкапсуляция)
Условие

Создайте базовый класс Animal, у которого есть:
защищённый атрибут _name
конструктор __init__(name)
метод speak(), который возвращает строку "Animal sound"

Создайте дочерний класс Dog, который:
наследуется от Animal
переопределяет метод speak()
возвращает строку "Woof"

Создайте объект Animal и объект Dog
и выведите результат работы метода speak().

Пример ожидаемого поведения
a = Animal("Unknown")
d = Dog("Rex")

print(a.speak()) # Animal sound
print(d.speak()) # Woof
```

```python

class Animal:
    def __init__(self, name):
        self._name = name  # защищённый атрибут

    def speak(self):
        return "Animal sound"


class Dog(Animal):
    def speak(self):
        return "Woof"


# создание объектов
a = Animal("Unknown")
d = Dog("Rex")

# вывод результатов
print(a.speak())  # Animal sound
print(d.speak())     # Woof
```

```python

class Animal:
    def __init__(self, name):
        self._name = name
    def speak(self):
        return "Animal sound"

class Dog(Animal):
    def speak(self):
        return "Woof"

a = Animal("Unknown")
d = Dog("Rex")

print(a.speak())
print(d.speak())   
```

```python
#задание 2
Задание 2. Функции
Условие
Напишите функцию count_words(text), которая:
принимает строку
считает количество слов
возвращает это количество

Пример: count_words("Python is very cool")
Ожидаемый результат: 4
```

```python
def count_words(text):
    return len(text.split())


# пример использования
print(count_words("Python is very cool"))  # 4
```

```python
def count_words(text):
    return len(text.split())

print(count_words("Python is very cool"))
```

```python

Задание 3. Объединённое
Условие

Создайте класс TextAnalyzer, у которого есть:
атрибут text
метод analyze(), который использует функцию count_words
и возвращает количество слов в тексте

Пример
analyzer = TextAnalyzer("I love Python")
print(analyzer.analyze()) # 3
```

```python
def count_words(text):
    return len(text.split())

class TextAnalyzer:
    def __init__(self, text):
        self.text = text

    def analyze(self):
        return count_words(self.text)


# пример
analyzer = TextAnalyzer("I love Python")
print(analyzer.analyze())  # 3
```

```python
class TextAnalyzer:
    def __init__(self, text):
        self.text = text

    def analyze(self):
        return len(self.text.split())


# пример
print(TextAnalyzer("I love Python").analyze())  # 3
```

```python
def count_words(text):
    return len(text.split())

class TextAnalyzer:
    def __init__(self, text):
        self.text = text
    def analyze(self):
        return count_words(self.text)

analyzer = TextAnalyzer("I love Python")
print(analyzer.analyze())
```
