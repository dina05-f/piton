```python
%%writefile bank_analytycs/metrics.py

def total_balance(accounts: list[dict])-> float:
    return sum(acc["balance"] for acc in accounts)

def count_rich(accounts: list[dict], thresholdL float)-> int:
    return sum(1 for axx in accounts if acc["balance"]>threshold)
```

```python
from bank_analytycs.metrics import total_balance, count_rich

accounts = [
    {"client": "Aijan", "balance": 1200},
    {"client": "Nurs", "balance": 500},
    {"client": "Sezim", "balance": 3000},
]

print("Total", total_balance(accounts))
print("Rich count (>1000):", count_rich(accounts, 1000))
```
