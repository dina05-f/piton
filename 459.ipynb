```python
def read_file(path):
    with open(path, "r", encoding="utf-8") as file:
        for line in file:
            print(line.strip())

read_file("clients.txt")
```

```python
def add(a, b):
    return a+b

print(add(5, 6))
```

```python
def add(a, b=8):#именованный
    return a+b

print(add(5, 6))
print(add(5))
```

```python
*args - позиционные аргументы

**kwargs - это набор именновых аргументов, к-е Python складывает в словарь
```

```python
def show_args(*args):
    print(args)

show_args(1, 2, 3, 4, 5)
```

```python
def sum_numbers(*args):
    total=0
    for num in args:
        total+=num
    return total
print(sum_numbers(1, 2, 3, 4, 5, 44))
```

```python
def sum_numbers(*args):
    return sum(args)
print(sum_numbers(1, 2, 3, 4, 5, 44))
```

```python
def show_kwargs(**kwargs):
    print(kwargs)

print(show_kwargs(name="Aizhan", age=28, city="Bishkek", job="Frontented developer"))
```

```python
def log_event(**data):
    for key, value in data.items():
        print(f'{key}: {value}')

log_event(user="Dina", action="login", status="ok")
```

```python
def demo(*args, **kwargs):
    print('args', args)
    print('kwargs', kwargs)

demo(1, 2, 3, name="Aizhan", age=28)
```

```python
def countdown(n):
    if n==0:
        return
    print(n)
    countdown(n-1)
countdown(10)
```

```python
def factorial(n):
    if n ==1:
        return 1
    return n*factorial(n-1)

factorial(5)
```
